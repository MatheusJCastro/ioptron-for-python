# Using the Driver - Basic Examples

Importing the driver:

```python
from ioptron_driver import Ioptron
import time
```

Selecting the port and initializing serial:

```python
#port = "/dev/ttyUSB0"  # Linux like
port = "COM2"           # Windows like

mount = Ioptron(port=port)
mount.init_serial(timedate=True)
```

The `timedate` variable is used to set or not the mount to the current time and date.

- In case `false` no changes are made. 
- In case `true` the driver will get time-date from the system and apply it to the mount.

Some example commands:

Commands without `exec` does not run and output the command listed. To run the command, add the `exec` as in the examples below.

```python
# Get Mount name
print(mount.device_name)

# Change move rate: 
print(mount.moving_rate(9), mount.exec.moving_rate(9))

# Move east for 5 sec
print(mount.move_east, mount.exec.move_east)
time.sleep(5)

# Then west but slower
print(mount.moving_rate(8), mount.exec.moving_rate(8))
print(mount.move_west, mount.exec.move_west)
time.sleep(5)

# Than stops
print(mount.stop_motion, mount.exec.stop_motion)
time.sleep(5)

# Than go to zero position
print(mount.go_home, mount.exec.go_home)

# Send direct commands, without the wrapper
# Note that commands here does not include : or #
print(mount.send_cmd("MountInfo"))
# Equivalent to mount.mount_model
```

Closing the serial communication:

```python
mount.close_serial()
```
